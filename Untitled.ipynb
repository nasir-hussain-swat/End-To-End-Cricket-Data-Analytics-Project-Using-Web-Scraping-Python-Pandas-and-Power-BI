```python
#importing libraries
import pandas as pd
import json
```

```python
with open("t20_json_files/t20_wc_match_results.json") as f:
    data=json.load(f)
df_match=pd.DataFrame(data[0]["matchSummary"])
df_match.head()
    
```

```python
#renaming column
df_match.rename({'scorecard':'matchId'},axis=1,inplace=True)
df_match.head()
```

```python
#creating another column to relate to another table later
match_ids_dict={}

for index,row in df_match.iterrows():
    key1=row['team1'] + ' VS ' + row['team2']
    key2=row['team2'] + ' VS ' + row['team1']

    match_ids_dict[key1]=row["matchId"]
    match_ids_dict[key2]=row["matchId"]

match_ids_dict
```

```python

match_ids_dict["Namibia VS Sri Lanka"]
```

```python
match_ids_dict["Pakistan VS England"]
```

```python
with open("t20_json_files/t20_wc_batting_summary.json") as f:
    data=json.load(f)
    all_rec=[]

    for rec in data:
        all_rec.extend(rec["battingSummary"])
        
```

```python
df_batting=pd.DataFrame(all_rec)
df_batting.head()
```

```python
#creating new column from the dismissal column
df_batting["out/notout"]=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "notout")
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head()
```

```python
df_batting["match_Id"]=df_batting["match"].map(match_ids_dict)
df_batting.drop["match_Id"]
```
