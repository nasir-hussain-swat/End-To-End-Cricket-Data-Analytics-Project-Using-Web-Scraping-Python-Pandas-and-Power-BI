## batting summary

```python
with open("t20_json_files/t20_wc_batting_summary.json") as f:
    data=json.load(f)
    all_rec=[]

    for rec in data:
        all_rec.extend(rec["battingSummary"])
        
```

```python
df_batting=pd.DataFrame(all_rec)
df_batting.head()
```

```python
#creating new column from the dismissal column
df_batting["out/notout"]=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "notout")
df_batting.head()
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head()
```
